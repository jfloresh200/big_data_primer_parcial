```python
#Ejercicio 7: Promedio de calificaciones
#• Defina o solicite 5 calificaciones.
#• Calcule la suma total.
#• Calcule el promedio.
#• Indique si el estudiante aprobó o reprobó.
#• Considere aprobado si el promedio es mayor o igual a 70.


calificaciones = [85.5, 90.0, 65.0, 72.5, 88.0]


suma_total = sum(calificaciones)


promedio = suma_total / len(calificaciones)


print("----Calificaciones----")
print(f"Notas: {calificaciones}")
print(f"Suma Total: {suma_total:.2f}")
print(f"Promedio Total: {promedio:.2f}")
print("-" * 33)

if promedio >= 70:
    print("Estado del Estudiante: APROVADO")
else:
    print("Estado del Estudiante: REPROBADO")
```

```python
#Ejercicio 8: Lista de números positivos y negativos
#• Defina o solicite 10 números.
#• Cuente cuántos son positivos, negativos y ceros.
#• Muestre un resumen final.

#Definicion de los numeros
numeros = [0, 0, 50, -45, 64, 75, -32, -85, 49, 41]

#Contadores de numeros
positivos = 0
negativos = 0
ceros = 0

#Ciclo for para contar los numeros positvos, negativos y ceros
for num in numeros:
    if num > 0:
        positivos += 1
    elif num < 0:
        negativos += 1
    else:
        ceros += 1

#Imprime los resultados finales
print("=== RESUMEN DE CLASIFICACIÓN ===")
print(f"Lista de números evaluados: {numeros}")
print("-" * 33)
print(f"Cantidad de números Positivos: {positivos}")
print(f"Cantidad de números Negativos: {negativos}")
print(f"Cantidad de números Ceros:     {ceros}")
print("================================")
```

```python
#Ejercicio 9: Mayor y menor de una lista
#• Cree una lista con varios números enteros.
#• Encuentre el número mayor.
#• Encuentre el número menor.
#• Calcule la diferencia entre ambos.
#• Explique el uso de max() y min().

# 1. Crear una lista con varios números enteros predefinidos
valores = [42, 17, 89, 4, 55, 23, 71, 12, 95, 31]

# 2. e 3. Encontrar el mayor y el menor usando las funciones integradas de Python
numero_mayor = max(valores)
numero_menor = min(valores)

# 4. Calcule la diferencia entre ambos (Mayor menos Menor)
diferencia = numero_mayor - numero_menor

# Mostrar el reporte de resultados en pantalla
print("=== ANÁLISIS DE EXTREMOS ===")
print(f"Lista de valores: {valores}")
print("-" * 30)
print(f"Número Mayor encontrado: {numero_mayor}")
print(f"Número Menor encontrado: {numero_menor}")
print(f"Diferencia (Mayor - Menor): {diferencia}")
print("============================")
```

```python
inventario = [
    {"codigo": "P001", "nombre": "Laptop", "precio": 15000, "cantidad": 5},
    {"codigo": "P002", "nombre": "Mouse", "precio": 350, "cantidad": 20},
    {"codigo": "P003", "nombre": "Teclado", "precio": 800, "cantidad": 10},
    {"codigo": "P004", "nombre": "Monitor", "precio": 4500, "cantidad": 7}
]

print("=== PRODUCTOS DISPONIBLES ===")

for producto in inventario:
    print(producto)

codigo_buscar = "P002"

print("\n=== BÚSQUEDA DE PRODUCTO ===")

for producto in inventario:
    if producto["codigo"] == codigo_buscar:
        print("Producto encontrado:")
        print(producto)

valor_total = 0

for producto in inventario:
    subtotal = producto["precio"] * producto["cantidad"]
    valor_total += subtotal

print("\n=== VALOR TOTAL DEL INVENTARIO ===")
print("Lps.", valor_total)
```

```python
ventas = [
    {"cliente": "Juan", "producto": "Laptop", "cantidad": 1, "precio_unitario": 15000},
    {"cliente": "Maria", "producto": "Mouse", "cantidad": 3, "precio_unitario": 350},
    {"cliente": "Carlos", "producto": "Monitor", "cantidad": 2, "precio_unitario": 4500},
    {"cliente": "Ana", "producto": "Teclado", "cantidad": 4, "precio_unitario": 800}
]

print("=== TOTAL POR VENTA ===")

total_general = 0
cliente_mayor = ""
compra_mayor = 0

for venta in ventas:

    total_venta = venta["cantidad"] * venta["precio_unitario"]

    print(
        venta["cliente"],
        "-",
        venta["producto"],
        "- Total: Lps.",
        total_venta
    )

    total_general += total_venta

    if total_venta > compra_mayor:
        compra_mayor = total_venta
        cliente_mayor = venta["cliente"]

print("\n=== TOTAL GENERAL VENDIDO ===")
print("Lps.", total_general)

print("\n=== CLIENTE CON MAYOR COMPRA ===")
print(cliente_mayor, "- Lps.", compra_mayor)
```

```python
# ejercicio 11
def calcular_descuento(precio_original, porcentaje_descuento):
    """
    Parámetros que entran:
      precio_original      → float/int  (ej: 500.00)
      porcentaje_descuento → float/int  (ej: 20  → 20%)
    Dato que retorna:
      precio_final         → float      (precio con descuento aplicado)
    """
    monto_descuento = precio_original * (porcentaje_descuento / 100)
    precio_final    = precio_original - monto_descuento
    return precio_final

# Prueba de la función
precio     = 500.00
descuento  = 20
final      = calcular_descuento(precio, descuento)

print(f"Precio original  : L. {precio:.2f}")
print(f"Descuento        : {descuento}%")
print(f"Precio final     : L. {final:.2f}")
```

```python
#ejercicio12
def validar_contrasena(contrasena):
    tiene_longitud  = len(contrasena) >= 8
    tiene_numero    = any(c.isdigit()  for c in contrasena)
    tiene_mayuscula = any(c.isupper()  for c in contrasena)
    tiene_minuscula = any(c.islower()  for c in contrasena)

    # Mostrar resultado de cada condición
    print(f"  Mínimo 8 caracteres   : {'yes' if tiene_longitud  else 'no'}")
    print(f"  Contiene número       : {'yes' if tiene_numero    else 'no'}")
    print(f"  Contiene mayúscula    : {'yes' if tiene_mayuscula else 'no'}")
    print(f"  Contiene minúscula    : {'yes' if tiene_minuscula else 'no'}")

    if tiene_longitud and tiene_numero and tiene_mayuscula and tiene_minuscula:
        return "yesContraseña VÁLIDA"
    else:
        return "no  Contraseña INVÁLIDA"

# Pruebas
for pwd in ["Segura1!", "débil", "SINMINUSCULA8"]:
    print(f'\nContraseña: "{pwd}"')
    print(validar_contrasena(pwd))
```
