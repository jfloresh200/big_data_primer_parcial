```python
# Ejer 19 
import pandas as pd


datos = {
    "Cliente": ["Juan", "María", "Carlos", "Ana", "Pedro", "Lucía"],
    "Monto solicitado": [45000, 60000, 30000, 80000, 55000, 40000],
    "Plazo (meses)": [12, 24, 18, 36, 24, 12],
    "Ingreso mensual": [15000, 25000, 12000, 30000, 22000, 18000],
    "Estado": ["Aprobado", "Rechazado", "Aprobado", "Aprobado", "Pendiente", "Rechazado"]
}

prestamos = pd.DataFrame(datos)

print("Lista de solicitudes de préstamo")
print(prestamos)


promedio = prestamos["Monto solicitado"].mean()

print("\nPromedio del monto solicitado:")
print(promedio)


print("\nCantidad de solicitudes por estado:")
print(prestamos["Estado"].value_counts())


print("\nPréstamos aprobados:")
print(prestamos[prestamos["Estado"] == "Aprobado"])


print("\nClientes con préstamos mayores a 50,000:")
print(prestamos[prestamos["Monto solicitado"] > 50000])

print("\nPromedio del monto solicitado según el estado:")
print(prestamos.groupby("Estado")["Monto solicitado"].mean())
```

```python
import pandas as pd


data = {
    "Nombre": [
        "Ana García", "Luis Martínez", "María López", "Carlos Pérez",
        "Sofía Hernández", "Jorge Ramírez", "Daniela Torres", "Andrés Flores",
        "Valeria Morales", "Ricardo Gómez"
    ],
    "Departamento": [
        "Ventas", "TI", "RRHH", "TI",
        "Ventas", "Finanzas", "RRHH", "Finanzas",
        "TI", "Ventas"
    ],
    "Cargo": [
        "Ejecutivo de Ventas", "Desarrollador", "Analista RRHH", "Arquitecto de Software",
        "Gerente de Ventas", "Contador", "Jefe de RRHH", "Analista Financiero",
        "Desarrollador Senior", "Asesor Comercial"
    ],
    "Salario": [
        35000, 52000, 40000, 75000,
        68000, 47000, 61000, 43000,
        70000, 32000
    ],
    "AniosExperiencia": [3, 6, 4, 10, 8, 5, 9, 2, 11, 1],
    "Ciudad": [
        "Bogotá", "Medellín", "Bogotá", "Cali",
        "Medellín", "Bogotá", "Cali", "Barranquilla",
        "Medellín", "Bogotá"
    ]
}

df = pd.DataFrame(data)

#  Primeros registros
print("=" * 60)
print("PRIMEROS REGISTROS")
print("=" * 60)
print(df.head())

#  Estadísticas generales de salarios
print("\n" + "=" * 60)
print("ESTADÍSTICAS GENERALES DE SALARIOS")
print("=" * 60)
print(df["Salario"].describe())

#  Salario promedio por departamento
print("\n" + "=" * 60)
print("SALARIO PROMEDIO POR DEPARTAMENTO")
print("=" * 60)
print(df.groupby("Departamento")["Salario"].mean())

#  Empleado con salario más alto
print("\n" + "=" * 60)
print("EMPLEADO CON SALARIO MÁS ALTO")
print("=" * 60)
print(df.loc[df["Salario"].idxmax()])

#  Empleados con más de 5 años de experiencia
print("\n" + "=" * 60)
print("EMPLEADOS CON MÁS DE 5 AÑOS DE EXPERIENCIA")
print("=" * 60)
print(df[df["AniosExperiencia"] > 5])

#  Ordenar empleados por salario
print("\n" + "=" * 60)
print("EMPLEADOS ORDENADOS POR SALARIO (MAYOR A MENOR)")
print("=" * 60)
print(df.sort_values("Salario", ascending=False))

#  Columna NivelExperiencia
def nivel_experiencia(anios):
    if anios <= 3:
        return "Junior"
    elif anios <= 6:
        return "Intermedio"
    else:
        return "Senior"

df["NivelExperiencia"] = df["AniosExperiencia"].apply(nivel_experiencia)

print("\n" + "=" * 60)
print("DATAFRAME CON COLUMNA NIVELEXPERIENCIA")
print("=" * 60)
print(df[["Nombre", "AniosExperiencia", "NivelExperiencia"]])
```

```python
import pandas as pd

# 1. Se crea un diccionario con los datos predefinidos de los 10 estudiantes
datos = {
    "nombre": ["Ana", "Luis", "Carlos", "Sofia", "Diego", "Maria", "Juan", "Elena", "Pedro", "Laura"],
    "carrera": ["Sistemas", "Industrial", "Civil", "Sistemas", "Electrica", "Industrial", "Civil", "Sistemas", "Mecanica", "Industrial"],
    "edad": [21, 23, 20, 22, 24, 21, 22, 20, 23, 22],
    "promedio": [85.5, 74.0, 92.3, 68.5, 81.0, 88.5, 79.0, 95.0, 62.0, 83.4],
    "ciudad": ["Bogota", "Medellin", "Cali", "Bogota", "Barranquilla", "Medellin", "Cali", "Bogota", "Bucaramanga", "Cartagena"]
}

# Convertir el diccionario en un DataFrame de Pandas
df_estudiantes = pd.DataFrame(datos)

# 2. Muestra las primeras 5 filas usando .head()
print("=== PRIMERAS 5 FILAS DEL DATAFRAME ===")
print(df_estudiantes.head())
print("\n" + "="*50 + "\n")

# 3. Muestre la informacion general del DataFrame usando .info()
print("=== INFORMACION GENERAL DEL DATAFRAME ===")
df_estudiantes.info()
print("\n" + "="*50 + "\n")

# 4. Calcule el promedio general de todas las calificaciones
promedio_general = df_estudiantes["promedio"].mean()
print(f"Promedio general de todos los estudiantes: {promedio_general:.2f}")
print("\n" + "="*50 + "\n")

# 5. Muestra estudiantes con promedio mayor o igual a 80 
print("=== ESTUDIANTES CON PROMEDIO MAYOR O IGUAL A 80 ===")
filtrados = df_estudiantes[df_estudiantes["promedio"] >= 80]
print(filtrados)
print("\n" + "="*50 + "\n")

# 6. Ordena estudiantes por promedio de mayor a menor
print("=== ESTUDIANTES ORDENADOS DE MAYOR A MENOR PROMEDIO ===")
ordenados = df_estudiantes.sort_values(by="promedio", ascending=False)
print(ordenados)
```
